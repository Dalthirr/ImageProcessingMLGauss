# Budowanie modelu CNN przy użyciu interfejsu Keras

### Pobieramy potrzebne biblioteki

```python
import numpy as np
import pandas as pd
import os
from sklearn.utils import shuffle
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import cv2
```

### Pobieramy zbiór danych treningowych

```python
# Mamy 6 podfolderów (sklasyfikowane),
# więc napiszmy prostą funkcję do pobierania obrazków wraz z etykietami


def wczytaj_obrazki(sciezka):

    # Tworzymy puste listy na przechowywanie obrazów i etykiet
    Obrazy = []
    Etykiety = []
    # 0 - buildings
    # 1 - forest
    # 2 - glacier
    # 3 - mountain
    # 4 - sea
    # 5 - street

    # Inicjujemy iterator po etykietach
    etykieta = 0

    for x in os.listdir(sciezka):  # Jesteśmy w głównym folderze
        # Klasyfikujemy eytkiety zgodnie z pow kodowaniem ustalonym przez nas
        if x == 'buildings':
            etykieta = 0
        elif x == 'forest':
            etykieta = 1
        elif x == 'glacier':
            etykieta = 2
        elif x == 'mountain':
            etykieta = 3
        elif x == 'sea':
            etykieta = 4
        elif x == 'street':
            etykieta = 5

        # Z każdego z folderów wyciągniemy teraz po kolei obrazek
        for plik in os.listdir(sciezka + x):
            # Wczytujemy obraz
            obraz = cv2.imread(sciezka + x + r'/' + plik)
            # Zdarzają się inne rozmiary obrazu, musimy go przeskalować
            obraz = cv2.resize(obraz, (150, 150))
            # Dopisujemy do listy obrazów
            Obrazy.append(obraz)
            # Zapisujemy etykietę do listy etykiet
            Etykiety.append(etykieta)
    # Mieszamy zbiór danych, który właśnie otrzymaliśmy
    return shuffle(Obrazy, Etykiety, random_state=42)


def wczytaj_obrazki_pred(sciezka):

    # Tworzymy puste listy na przechowywanie obrazów
    Obrazy = []
    for plik in os.listdir(sciezka):
            # Wczytujemy obraz
            obraz = cv2.imread(sciezka + r'/' + plik)
            # Zdarzają się inne rozmiary obrazu, musimy go przeskalować
            obraz = cv2.resize(obraz, (150, 150))
            # Dopisujemy do listy obrazów
            Obrazy.append(obraz)
            # Zapisujemy etykietę do listy etykiet
    # Mieszamy zbiór danych, który właśnie otrzymaliśmy
    return shuffle(Obrazy, random_state=42)


# Utwórzmy jeszcze funkcję, która zwróci nam odpowiadającą etykietę danej cyfrze
def slownik_etykiety(cyfra):

    etykiety = {0: 'buildings', 1: 'forest', 2: 'glacier',
                3: 'mountain', 4: 'sea', 5: 'street'
                }
    return etykiety[cyfra]
```

### Mamy już potrzebną funkcję do importowania danych, wczytajmy zatem zbiór treningowy

```python
obrazy_treningowe, etykiety_treningowe = wczytaj_obrazki('../Model CNN/seg_train/')

obrazy_treningowe = np.array(obrazy_treningowe) 
etykiety_treningowe = np.array(etykiety_treningowe)
```

### Zerknijmy na kilka losowych obrazków

```python
# Zerknijmy na 16 obrazków
figure, ax = plt.subplots(4, 4)
figure.subplots_adjust(0, 0, 3, 3)
for i in range(0, 4, 1):
    for j in range(0, 4, 1):
        # Wybieramy losowy indeks z całego zbioru
        losowy_indeks = np.random.randint(0, len(obrazy_treningowe))
        # Wrzucamy losowy obrazek
        ax[i, j].imshow(obrazy_treningowe[losowy_indeks])
        # Tytuł będzie etykietą
        ax[i, j].set_title(slownik_etykiety(etykiety_treningowe[losowy_indeks]))
        # Ukrywamy osie
        ax[i, j].axis('off')
```

### Czas na sprawdzenie skuteczności modelu na zbioru testowym

```python
# Wczytujemy dane
obrazy_testowe, etykiety_testowe = wczytaj_obrazki('../Model CNN/seg_test/')
obrazy_testowe = np.array(obrazy_testowe)
etykiety_testowe = np.array(etykiety_testowe)
# Testujemy
model.evaluate(obrazy_testowe, etykiety_testowe, batch_size=8, verbose=1)
```

### Zerknijmy na obrazki, którym nasz model przypisał błędne etykiety

```python
pred_prob = model.predict(obrazy_testowe, batch_size=8)
pred_klasy = pred_prob.argmax(axis=-1)
```

```python
nietrafione = np.nonzero(pred_klasy != etykiety_testowe)
nietrafione = np.array(nietrafione).squeeze(axis=0)
nietrafione[:10]
```

##### Otrzymaliśmy indeksy obrazów ze zbioru treningowego, w których nie trafiliśmy z predykcją - zerknijmy na kilka z nich

```python
wykres, ax = plt.subplots(3, 3, figsize=(20, 20))

for i in range(0, 3, 1):
    for j in range(0, 3, 1):
        # Losujemy indeks z listy nietrafionych
        indeks = np.random.choice(nietrafione)
        # Wybieramy obrazek
        obrazek_nietrafiony = obrazy_testowe[indeks]
        # Wrzucamy losowy obrazek
        ax[i, j].imshow(obrazek_nietrafiony)
        # Tytuł będzie etykietą
        ax[i, j].set_title('R: {0}, P: {1}'.format(slownik_etykiety(etykiety_testowe[indeks]),
                                                   slownik_etykiety(pred_klasy[indeks]))
                                                   )
        # Ukrywamy osie
        ax[i, j].axis('off')
plt.show()
# R - etykieta testowa
# P - etykieta predykowana przez model
```
